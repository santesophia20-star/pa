#Entrada
 Média de 3 números

Crie um algoritmo que:
	•	Leia três números
	•	Calcule e mostre a média

```python
# Média de 3 números

# --- ENTRADA ---
# Lendo os três números do usuário
n1 = float(input("Digite o 1º número: "))
n2 = float(input("Digite o 2º número: "))
n3 = float(input("Digite o 3º número: "))

# --- PROCESSAMENTO ---
# Calculando a média aritmética
media = (n1 + n2 + n3) / 3

# --- SAÍDA ---
# Exibindo o resultado final formatado
print("-" * 30)
print(f"Os números digitados foram: {n1}, {n2} e {n3}")
print(f"A média final calculada é: {media:.2f}")
print("-" * 30)
```

 Divisão inteira e resto

Faça um programa que:
	•	Leia um número inteiro
	•	Mostre:
	•	O dobro
	•	A metade
	•	O resto da divisão por 3

```python
# --- ENTRADA ---
numero = int(input("Digite um número inteiro: "))

# --- PROCESSAMENTO ---
dobro = numero * 2
metade = numero / 2
resto = numero % 3

# --- SAÍDA ---
print("-" * 30)
print(f"Número digitado: {numero}")
print(f"O dobro é: {dobro}")
print(f"A metade é: {metade}")
print(f"O resto da divisão por 3 é: {resto}")
print("-" * 30)
```

Conversão de temperatura

Desenvolva um programa que:
	•	Leia uma temperatura em Celsius
	•	Converta para Fahrenheit

```python
# --- ENTRADA ---
celsius = float(input("Digite a temperatura em Celsius (°C): "))

# --- PROCESSAMENTO ---
# A fórmula multiplica por 1.8 e soma 32
fahrenheit = (celsius * 1.8) + 32

# --- SAÍDA ---
print("-" * 35)
print(f"Temperatura em Celsius: {celsius}°C")
print(f"Temperatura em Fahrenheit: {fahrenheit:.1f}°F")
print("-" * 35)
```

Cálculo de salário

Faça um algoritmo que:
	•	Leia o salário de um funcionário
	•	Calcule um aumento de 10%
	•	Mostre o novo salário

```python
# --- ENTRADA ---
salario_atual = float(input("Digite o salário atual do funcionário (R$): "))

# --- PROCESSAMENTO ---
# Opção 1: Calcular o aumento e somar (aumento = salario_atual * 0.10)
# Opção 2: Multiplicar diretamente por 1.10 (forma mais rápida)
novo_salario = salario_atual * 1.10

# --- SAÍDA ---
print("-" * 35)
print(f"Salário Anterior: R$ {salario_atual:.2f}")
print(f"Aumento aplicado: 10%")
print(f"Novo Salário: R$ {novo_salario:.2f}")
print("-" * 35)
```

Área de um retângulo

Crie um programa que:
	•	Leia a base e a altura
	•	Calcule a área


```python
# --- ENTRADA ---
base = float(input("Digite o valor da base: "))
altura = float(input("Digite o valor da altura: "))

# --- PROCESSAMENTO ---
# A área é o produto da base pela altura
area = base * altura

# --- SAÍDA ---
print("-" * 30)
print(f"Base: {base}")
print(f"Altura: {altura}")
print(f"A área do retângulo é: {area:.2f}")
print("-" * 30)
```

Consumo de combustível

Faça um programa que:
	•	Leia a distância percorrida (km)
	•	Leia o combustível gasto (litros)
	•	Calcule o consumo médio (km/l)

```python
# --- ENTRADA ---
distancia = float(input("Digite a distância percorrida (km): "))
combustivel = float(input("Digite a quantidade de combustível gasta (litros): "))

# --- PROCESSAMENTO ---
# O consumo médio é a distância dividida pelo combustível
consumo_medio = distancia / combustivel

# --- SAÍDA ---
print("-" * 40)
print(f"Distância: {distancia} km")
print(f"Combustível: {combustivel} litros")
print(f"O consumo médio do veículo é: {consumo_medio:.2f} km/l")
print("-" * 40)
```
