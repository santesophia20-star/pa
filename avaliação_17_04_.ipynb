#questão 1
 Par ou ímpar
Peça um número ao usuário e informe se ele é par ou ímpar.

```python
# Solicita a entrada do usuário
numero = int(input("Digite um número inteiro: "))

# Verifica a paridade
if numero % 2 == 0:
    print(f"O número {numero} é PAR.")
else:
    print(f"O número {numero} é ÍMPAR.")
```

#questão 7
Calculadora simples
Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o
resultado.

```python
# Solicitando os números ao usuário (convertendo para float para permitir decimais)
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Solicitando a operação
operacao = input("Escolha a operação (+, -, *, /): ")

# Estrutura if/elif para cálculo
if operacao == "+":
    resultado = num1 + num2
    print(f"Resultado: {num1} + {num2} = {resultado}")
elif operacao == "-":
    resultado = num1 - num2
    print(f"Resultado: {num1} - {num2} = {resultado}")
elif operacao == "*":
    resultado = num1 * num2
    print(f"Resultado: {num1} * {num2} = {resultado}")
elif operacao == "/":
    # Verificação para evitar divisão por zero
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado: {num1} / {num2} = {resultado}")
    else:
        print("Erro: Divisão por zero não permitida.")
else:
    print("Operação inválida!")
```

#questão 8
Dia da semana
Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

```python
# Solicita um número ao usuário
try:
    numero = int(input("Digite um número de 1 a 7: "))

    # Lista com os dias da semana
    dias = [
        "Domingo",
        "Segunda-feira",
        "Terça-feira",
        "Quarta-feira",
        "Quinta-feira",
        "Sexta-feira",
        "Sábado"
    ]

    # Verifica se o número está no intervalo correto
    if 1 <= numero <= 7:
        # Subtraímos 1 porque as listas em Python começam no índice 0
        print(f"O dia correspondente é: {dias[numero - 1]}")
    else:
        print("Erro: Por favor, digite um número entre 1 e 7.")

except ValueError:
    print("Erro: Entrada inválida. Digite apenas números inteiros.")
```

#questão 9
Ano bissexto
Solicite um ano e informe se ele é bissexto ou não.

```python
# Solicita o ano ao usuário
ano = int(input("Digite o ano que deseja verificar: "))

# Verifica se o ano é bissexto
# Regra: Divisível por 4 E (não divisível por 100 OU divisível por 400)
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é BISSEXTO.")
else:
    print(f"O ano {ano} NÃO é bissexto.")
```

#questão 10
Desconto em compras
Peça o valor de uma compra:
 ≥ 100: 10% de desconto
 ≥ 50 e &lt; 100: 5% de desconto
 &lt; 50: sem desconto

```python
# Solicita o valor da compra
valor_compra = float(input("Digite o valor da compra: R$ "))

# Define as faixas de desconto
if valor_compra >= 100:
    desconto = 0.10  # 10%
elif valor_compra >= 50:
    desconto = 0.05  # 5%
else:
    desconto = 0.0   # Sem desconto

# Calcula o valor do desconto e o preço final
valor_desconto = valor_compra * desconto
preco_final = valor_compra - valor_desconto

# Exibe o resultado formatado
print("-" * 30)
print(f"Valor original: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {desconto*100:.0f}%")
print(f"Valor do desconto: R$ {valor_desconto:.2f}")
print(f"Preço final: R$ {preco_final:.2f}")
print("-" * 30)
```
