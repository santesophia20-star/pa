Divisão

#Entrada:

Dois números

#Processamento:

Dividir o primeiro pelo segundo (verificando divisão por zero)

#Saída:

Resultado da divisão ou mensagem de erro



```python
a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))

if b != 0:
    resultado = a / b
    print("Divisão:", resultado)
else:
    print("Erro: divisão por zero não é permitida.")
```

Média de 3 números

#Entrada:

Três números

#Processamento:

Somar os três números e dividir por 3

#Saída:

Resultado da média

```python
a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))
c = float(input("Digite o terceiro número: "))

media = (a + b + c) / 3

print("Média:", media)
```

Divisão inteira e resto

#Entrada:

Um número inteiro

#Processamento:

Calcular o dobro, metade e resto da divisão por 3

#Saída:

Resultados

```python
num = int(input("Digite um número inteiro: "))

dobro = num * 2
metade = num / 2
resto = num % 3

print("Dobro:", dobro)
print("Metade:", metade)
print("Resto da divisão por 3:", resto)
```

Conversão de temperatura

#Processamento:

Converter para Fahrenheit usando a fórmula F = (C * 9/5) + 32

Temperatura em Fahrenheit

```python
c = float(input("Digite a temperatura em Celsius: "))

f = (c * 9/5) + 32

print("Temperatura em Fahrenheit:", f)
```

Consumo de combustível

#Entrada:

Distância percorrida (km)
Combustível gasto (litros)

#Processamento:

Dividir a distância pelo combustível

#Saída:

Consumo médio (km/l)

```python
distancia = float(input("Digite a distância (km): "))
litros = float(input("Digite o combustível gasto (litros): "))

if litros != 0:
    consumo = distancia / litros
    print("Consumo médio:", consumo, "km/l")
else:
    print("Erro: combustível não pode ser zero.")
```
