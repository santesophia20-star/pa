```python
#3
#numeros pares, o usuario digita um numero e o programa imprimi os numeros pares

numero = int(input("digite um numero positivo:"))
print(f"Selecionando números pares até {numero}:")

for i in range(numero + 1):
    if i % 2 == 0:
        print(i)
```

```python
#4
#numeros impares, o usuario digita um numero e o programa imprimi os numeros impares

numero=int(input("digite um numero positivo"))
print(f"selecionar numeros impares que contem dentro de {numero}:")

for i in range(numero + 1):
    if i % 2 != 0:
        print(i)
```
